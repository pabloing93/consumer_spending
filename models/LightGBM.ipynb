```python
import pandas
import numpy
from sklearn.metrics import r2_score, mean_squared_error
import lightgbm as lgbm
```

```python
data_test = pandas.read_csv('../database/traffic_site_test.csv', sep=';')
data_test.head()
```

```python
data_train = pandas.read_csv('../database/traffic_site_train.csv', sep=';')
data_train.head()
```

```python
data_train.info()
```

```python
x_train = data_train.drop(columns='transactionRevenue')
y_train = data_train['transactionRevenue']

x_test = data_test.drop(columns='transactionRevenue')
y_test = data_test['transactionRevenue']
```

Entrenamos el modelo

```python
model = lgbm.LGBMRegressor(random_state=123)
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
predictions
```

Evaluamos el desempeÃ±o

```python
r2 = r2_score(y_test, predictions)
rmse = numpy.sqrt(mean_squared_error(y_test, predictions))

print(f'r2: {r2}')
print(f'rmse: {rmse/1000000}')
```
